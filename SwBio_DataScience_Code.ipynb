```python
import pandas as pd


#Loading in x3 stomatal aperture datasets from automated YOLO pipeline
ld1 = pd.read_csv("LD1_Automated_Width.csv")
ld1["Source"] = "LD1_Automated"

ld2 = pd.read_csv("LD2_Automated_Width.csv")
ld2["Source"] = "LD2_Automated"

ld3 = pd.read_csv("LD3_Automated_Width.csv")
ld3["Source"] = "LD3_Automated"

ld1.head()
ld2.head()
ld3.head()
```

```python
#Combining the x3 repeat datsets into one workable dataframe

ld_combined = pd.concat([ld1, ld2, ld3], ignore_index=True)

ld_combined.head()
```

```python
# Strip leading or trailing spaces from all column names
ld_combined.columns = ld_combined.columns.str.strip()

# Drop empty column if they exists
ld_combined = ld_combined.drop(columns=[""], errors="ignore")

ld_combined.columns
```

```python
#Ensures measurments and confidence columns are numeric and coerce invalid enteries intos NaN
numeric_cols = [
    "Pore Width",
    "Pore Length",
    "Pore Length Conf",
    "Aperture Confidence",
    "Detection Confidence",
    "Area"
]

for col in numeric_cols:
    ld_combined[col] = pd.to_numeric(ld_combined[col], errors="coerce")

ld_combined[numeric_cols].describe()
```

```python
#Etracts experimental metadata (ABA conc., genotype, plant replicate) from image filenames
parts = ld_combined["Image"].astype(str).str.split("_", expand=True)

ld_combined["ABAConc"] = parts[0]
ld_combined["Genotype"] = parts[1]
ld_combined["Plant"] = parts[2]

ld_combined[["Image", "ABAConc", "Genotype", "Plant"]].head()
```

```python
# Standardise classification labels by removing leading or trailing spaces

ld_combined["Classification"] = ld_combined["Classification"].str.strip()

ld_combined["Classification"].unique()
```

```python
#Applying confidence based quality control filtering 
#retains only high confidence detections for open stomata

mask = (
    (ld_combined["Detection Confidence"] > 0.6) &
    (
        (ld_combined["Classification"] == "closed") |
        (
            (ld_combined["Classification"] == "open") &
            (ld_combined["Aperture Confidence"] > 0.6)
        )
    )
)

ld_proc = ld_combined.loc[mask].copy()
ld_combined.shape, ld_proc.shape
```

```python
#Summarises raw vs QC-retained counts by classification

qc_table = (
    ld_combined
    .groupby("Classification")
    .size()
    .rename("Raw Count")
    .to_frame()
    .join(
        ld_proc
        .groupby("Classification")
        .size()
        .rename("QC Count")
    )
)

qc_table["% Kept"] = (qc_table["QC Count"] / qc_table["Raw Count"]) * 100

qc_table
```

```python
#Calculates and aggregates plant level summary stats for stomatal aperture measurements

ld_mean = (
    ld_proc
    .groupby(["Genotype", "ABAConc", "Plant", "Source"], as_index=False)
    .agg(
        n_width=("Pore Width", lambda x: x.dropna().shape[0]),
        n_length=("Pore Length", lambda x: x.dropna().shape[0]),
        std_width=("Pore Width", "std"),
        Width=("Pore Width", "mean"),
        Length=("Pore Length", "mean"),
    )
)

# Removing missing mean values
ld_mean = ld_mean.dropna(subset=["Width"])
ld_mean.head()
```

```python
#Prepares the dataframe for plotting and visualisation

import numpy as np
import matplotlib.pyplot as plt


plot_df = ld_mean[ld_mean["ABAConc"].isin(["0", "0.1", "1", "10"])].copy()

plot_df["Width_um"] = plot_df["Width"] * 0.248


plot_df["ABAConc"] = plot_df["ABAConc"].astype(str)
plot_df["Genotype"] = plot_df["Genotype"].astype(str)
```

```python

# Visualise plant level aperture widths across ABA concentrations and genotypes as boxplot

# Ordering ABA concentrations and genotypes
aba_levels = ["0", "0.1", "1", "10"]
genotypes = ["Col0", "156157"]

# Mapping internal genotype codes to display labels and colours
label_map = {
    "Col0": "Col-0",
    "156157": "mir156a/c mir157a/c",
}

color_map = {
    "Col0": "#E69F00",
    "156157": "#56B4E9",
}

# x positions for each ABA level
x_base = np.arange(len(aba_levels))
box_width = 0.35  

# Collect data for each box
box_data = []
box_positions = []
box_colors = []

for i, conc in enumerate(aba_levels):
    for j, geno in enumerate(genotypes):
        sub = plot_df[
            (plot_df["ABAConc"] == conc) &
            (plot_df["Genotype"] == geno)
        ]["Width_um"].dropna()

        if sub.empty:
            continue

        pos = x_base[i] + (j - 0.5) * box_width

        box_data.append(sub.values)
        box_positions.append(pos)
        box_colors.append(color_map[geno])

# Create figure
fig, ax = plt.subplots(figsize=(6, 6))

# Drawing boxplots
bp = ax.boxplot(
    box_data,
    positions=box_positions,
    widths=box_width * 0.8,
    patch_artist=True,
    showfliers=False
)

# Style boxes
for patch, color in zip(bp["boxes"], box_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.9)

# Style medians and whiskers
for median in bp["medians"]:
    median.set_color("black")
    median.set_linewidth(1.2)

for whisker in bp["whiskers"]:
    whisker.set_color("black")

for cap in bp["caps"]:
    cap.set_color("black")

# Overlaying jittered points for each group
rng = np.random.default_rng(42)

for pos, data, color in zip(box_positions, box_data, box_colors):
    x_jitter = pos + (rng.random(len(data)) - 0.5) * (box_width * 0.4)
    ax.scatter(
        x_jitter,
        data,
        s=25,
        edgecolor="black",
        linewidth=0.5,
        facecolor=color,
        alpha=0.9,
        zorder=3
    )

# Axis formatting
ax.set_xticks(x_base)
ax.set_xticklabels(aba_levels)
ax.set_xlabel("ABA Conc. (M)")
ax.set_ylabel("Mean Stomatal Aperture Width (m)")
ax.set_ylim(0, 5.2)
ax.set_xlim(-0.5, len(aba_levels) - 0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Legend
from matplotlib.patches import Patch

legend_handles = [
    Patch(facecolor=color_map[geno], edgecolor="black", label=label_map[geno])
    for geno in genotypes
]

ax.legend(
    handles=legend_handles,
    title="Genotype",
    frameon=False,
    loc="upper right"
)

plt.tight_layout()
plt.show()
```
